```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python

np.random.seed(58)
coefficients = np.random.randint(-10, 11, size=6)
polynomial = np.poly1d(coefficients)
x_values = np.linspace(-5, 5, 1000)
y_values = polynomial(x_values)
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Polynomial')
res_max = minimize(lambda x: -polynomial(x), 0, bounds=[(-5, 5)])
res_min = minimize(lambda x: polynomial(x), 0, bounds=[(-5, 5)])
plt.scatter([res_max.x[0], res_min.x[0]], [polynomial(res_max.x[0]), polynomial(res_min.x[0])], color='red')
plt.text(res_max.x[0], polynomial(res_max.x[0]), f'Max: ({res_max.x[0]:.2f}, {polynomial(res_max.x[0]):.2f})', fontsize=10, ha='right')
plt.text(res_min.x[0], polynomial(res_min.x[0]), f'Min: ({res_min.x[0]:.2f}, {polynomial(res_min.x[0]):.2f})', fontsize=10, ha='right')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of a 5th Degree Polynomial with Annotated Extrema')
plt.legend()
plt.grid(True)
plt.show()
```
